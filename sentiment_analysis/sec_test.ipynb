```python
url = "https://www.sec.gov/files/company_tickers.json"
headers = {"User-Agent": "Benjamin Boucher(benjamin.boucher-charest@polymtl.ca)"}

response = requests.get(url, headers=headers)
ticker_dict = {}
if response.status_code == 200:
    try:
        json_response = response.json()
        ticker_dict = {item['ticker']: item['cik_str'] for item in json_response.values()}
 
    except requests.exceptions.JSONDecodeError:
        print("Error: Response is not in JSON format.")
else:
    print(f"Request failed with status code: {response.status_code}")
```

```python
ticker = "MSFT"
ticker_CIK = str(ticker_dict[ticker])
num_zeros = 10 - len(str(ticker_dict[ticker]))
zero = '0' * num_zeros
url = f"https://data.sec.gov/api/xbrl/companyconcept/CIK{zero + ticker_CIK}/us-gaap/Assets.json"

headers = {
    "User-Agent": "Benjamin Boucher (benjamin.boucher-charest@polymtl.ca)"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()
    with open(f"AccountsPayableCurrent_{ticker}.txt", "w") as file:
        file.write(str(data))
else:
    print(f"Request failed with status code: {response.status_code}")
```

```python
import numpy as np

resp_data = data['units']['USD']

records = []
for entry in data['units']['USD']:
    record = (
        entry['end'], entry['val'], entry['accn'], entry['fy'],
        entry['fp'], entry['form'], entry['filed']
    )
    records.append(record)

dtype = [
    ('end', 'U10'),       
    ('value', 'i8'),      
    ('accn', 'U20'),      
    ('fy', 'i4'),         
    ('fp', 'U2'),         
    ('form', 'U4'),       
    ('filed', 'U10')      
]

np_table = np.array(records, dtype=dtype)
np_table =  np_table[np_table['form'] == '10-Q']
print(np_table)
```
